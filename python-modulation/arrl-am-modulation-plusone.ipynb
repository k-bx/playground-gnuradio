## Why 1+?

Let's see why exactly do we do the $1 + m(t)$

First, let's visualise $AM = c(t) \times [1 + m(t)] = [1 + M cos(2 \pi f_M t)] \times C sin(2 \pi f_C t)$

Begin with a simple sine wave for $c(t) = C sin(2 \pi f_C t)$

```python
time = np.arange(0, 0.02, 0.00001)
C = 3
f_C = 1800
print("> f_C: ", f_C)
ys = [C * np.sin(2 * np.pi * f_C * t) for t in time]
plot.plot(time, ys)
```

Now plot $m(t) = M cos(2 \pi f_M t)$

```python
M = 1
f_M = 80
print("f_M: ", f_M)
ys = [M * np.cos(2 * np.pi * f_M * t) for t in time]
plot.plot(time, ys)
```

Now plot the whole thing: $AM = c(t) \times [1 + m(t)] = [1 + M cos(2 \pi f_M t)] \times C sin(2 \pi f_C t)$

```python
ys = [(1 + M * np.cos(2 * np.pi * f_M * t)) * C * np.sin(2 * np.pi * f_C * t) for t in time]
plot.plot(time, ys)
```

Now, we'd like to see a similar modulation but without the +1 part.

$AM' = c(t) \times m(t) = M' cos(2 \pi f_{M'} t) \times C sin(2 \pi f_C t)$

```python
M_ = 1
f_M_ = 80
print("f_M: ", f_M_)
ys = [M_ * np.cos(2 * np.pi * f_M_ * t) * C * np.sin(2 * np.pi * f_C * t) for t in time]
plot.plot(time, ys)
```

## Modulation index

Let's decode these two terms:

- Modulation index = $M / C$
- Modulation percentage = $(M / C) × 100%$

Just as an experiment, try setting the $M$ constant to $M = C / 2$ and see, maybe they do mean $M$ to mean $M$ and $C$ to mean $C$ which would be wonderful after all?

```python
M_50p = C / 2
ys = [(M_50p + M_50p * np.cos(2 * np.pi * f_M * t)) * C * np.sin(2 * np.pi * f_C * t) for t in time]
plot.plot(time, ys)
```

Let's try and achieve an image-like result somehow and see if we can infer what they mean from that.

Yay, this is good enough for my own understanding! Let's try **overmodulation**:

```python
mod_index = 1.5
ys = [(1 + mod_index * M * np.cos(2 * np.pi * f_M * t)) * C * np.sin(2 * np.pi * f_C * t) for t in time]
plot.plot(time, ys)
```
